##DATA PREPRATION

```python
#CATEGORICAL (OBJECT DATA TYPES
category = df.select_dtypes(include='object')
category.head()
```

```python
##dummies
df_dummies = pd.get_dummies(category,drop_first=True)
df_dummies = df_dummies.astype('int')
```

```python
df_dummies
```

```python
#now we will remove the category1 columns
df = df.drop(list(category.columns),axis=1)
df
```

```python
###now we will append the category1 dummies (one-hot encoding)
df = pd.concat([df,df_dummies], axis=1)
```

```python
df
```

```python
df = df.drop(['instant','dteday'],axis=1,inplace=False)
df.head()
```

##BUILDING MODEL

```python
#FINAL DATASET LOOK LIKE,NOW WE WILL START PREDICTING USING LINEAR REGRESSION MODEL AND WLL SPLIT UP THE DATASET
df.head(2)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test = train_test_split(df,train_size=0.7,test_size=0.3,random_state=100)
```

```python
df_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
#now we will convert the col to scalar expect dummies col as they are already in 0-1 form

var = ['temp','hum','windspeed','casual','registered','cnt']
```

```python
df_train[var] = scalar.fit_transform(df_train[var])
```

```python
df_train.describe()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df_train.corr(),annot=True)
```

```python
y_train = df_train.pop('cnt')
x_train = df_train.drop(['casual','registered'],axis=1)
```

```python
x_train.head(2)
```

```python
np.asarray(df_train)
```

```python
import statsmodels.api as sm

x_train_lm = sm.add_constant(x_train)
lr= sm.OLS(y_train, x_train_lm).fit()
```

```python
x_train_lm
```

```python
lr.params.reset_index().rename(columns={0:'learning rate'})
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
lr.summary()
```

```python
#now we will reduce the dimension using feature selection process

from sklearn.feature_selection import RFE
```

```python
lm = LinearRegression()
rfe1 = RFE(lm,n_features_to_select = 15) # Changed the second argument to a keyword argument


rfe1.fit(x_train,y_train)
print(rfe1.support_)
print(rfe1.ranking_)
```

```python
col1 = x_train.columns[rfe1.support_]
col1
```

```python
x_train_rfe1 = x_train[col1]
```

```python
x_train_rfe1 = sm.add_constant(x_train_rfe1)
lm1 = sm.OLS(y_train,x_train_rfe1).fit()
lm1.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
a = x_train_rfe1.drop('const',axis=1)
```

```python
vif = pd.DataFrame()
vif['Features'] = a.columns
vif['VIF'] = [variance_inflation_factor(a.values,i) for i in range(a.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF',ascending=False)
vif
```

```python
lm = LinearRegression()
rfe2 = RFE(lm,n_features_to_select = 9) # Changed the second argument to a keyword argument


rfe2.fit(x_train,y_train)
print(rfe2.support_)
print(rfe2.ranking_)
```

```python
col2 = x_train.columns[rfe2.support_]
x_train_rfe2 = x_train[col2]

x_train_rfe2 = sm.add_constant(x_train_rfe2)
lm2 = sm.OLS(y_train,x_train_rfe2).fit()
lm2.summary()
```

```python
b = x_train_rfe2.drop('const',axis=1)
vif1 = pd.DataFrame()
vif1['Features'] = b.columns
vif1['VIF'] = [variance_inflation_factor(b.values,i) for i in range(b.shape[1])]
vif1['VIF'] = round(vif1['VIF'],2)
vif1 = vif1.sort_values(by='VIF',ascending=False)
vif1
```

```python
y_train_cnt = lm2.predict(x_train_rfe2)
```
